## Dataset Unificado

```python
# Partimos del dataset que contiene el log de visitas, al cual le agregamos informacion de los usuarios
ds = coupon_visit_train.merge(user_list, left_on='USER_ID_hash', right_on='USER_ID_hash', how='left')
```

```python
# vefiricamos no perdimos datos
print(len(coupon_visit_train))
print(len(ds))
```

```python
# Ahora sumamos la informacion de los cupones y verificamos cantidad de registros
ds = ds.merge(coupon_list, left_on='VIEW_COUPON_ID_hash', right_on='COUPON_ID_hash', how='left')
print(len(ds))
```

```python
# Veamos si el merge realizado esta correctamente bajo el supuesto de que VIEW_COUPON_ID_hash = COUPON_ID_hash
ds[ds['VIEW_COUPON_ID_hash'] == '34c48f84026e08355dc3bd19b427f09a'][['VIEW_COUPON_ID_hash' ,'COUPON_ID_hash']].head()
```

```python
# Vemos cantidad única de compras
ds.PURCHASEID_hash.nunique()
```

```python
# Informacion de las compras y dataset
print('Variables en dataset',coupon_detail_train.columns)
print('Cantidad de registros:', len(coupon_detail_train))
coupon_detail_train.head()
```

Los PURCHASEID_hash son unicos, identifican la compra,veamos si mapean con DS.

```python
# Chequeo de macheo
ds[ds['PURCHASEID_hash'] == 'c820a8882374a4e472f0984a8825893f'][['USER_ID_hash', 'COUPON_ID_hash']]
```

```python
# Realizamos el merge entre ds y coupon_detail_train con el campo PURCHASEID_hash
coupon_detail_list_columns = ['ITEM_COUNT', 'I_DATE','SMALL_AREA_NAME', 'PURCHASEID_hash']
ds = ds.merge(coupon_detail_train[coupon_detail_list_columns], left_on='PURCHASEID_hash',
              right_on='PURCHASEID_hash', how='left')
print('Cantidad de registros:', len(ds))
ds.columns
```

```python
# Depuramos columnas duplicadas en los merge o que no vamos a utilizar en el analisis
#to_drop = ['USER_ID_hash_y', 'COUPON_ID_hash_y', 'PURCHASEID_hash']
#ds = ds.drop(to_drop, axis='columns')
ds = ds.rename(columns={'I_DATE_y':'SALE_DATE',
                  'I_DATE_x': 'VIEW_DATE'})
ds.columns
```

```python
# Chequeamos informacion del dataset obtenido
ds.info()
```

```python
# Pasamos las fechas de string a datetime
ds['VIEW_DATE'] = pd.to_datetime(ds['VIEW_DATE'])
ds['WITHDRAW_DATE'] = pd.to_datetime(ds['WITHDRAW_DATE'])
ds['SALE_DATE'] = pd.to_datetime(ds['SALE_DATE'])
```

```python
#Le agregamos las coordenadas
col_pref_loc = ['PREF_NAME', 'LATITUDE', 'LONGITUDE']
ds = ds.merge(prefecture_locations[col_pref_loc], left_on='PREF_NAME', right_on='PREF_NAME',how='left')
print('Total registros:', len(ds))
```

```python
# Agregamos traducion de CAPSULE_TEXT
capsule = CAPSULE_TEXT_Translation[['CAPSULE_TEXT','English_Translation_ct']]
ds = ds.merge(capsule, left_on='CAPSULE_TEXT', right_on='CAPSULE_TEXT', how='left')
print('Total registros:',len(ds))
```

```python
# Modifiquemos los nombres de las variables para poder trabajar mas facil
ds.rename(columns={'PURCHASE_FLG':'compra',
                  'VIEW_DATE':'fecha_visita',
                  'PAGE_SERIAL':'page_serial',
                  'REFERRER_hash': 'referrer_hash',
                  'VIEW_COUPON_ID_hash':'ID_cupon_visto',
                  'USER_ID_hash':'ID_usuario',
                  'SESSION_ID_hash':'ID_sesion',
                  'PURCHASEID_hash':'ID_compra',
                  'REG_DATE':'fecha_reg_usuario',
                  'SEX_ID':'sexo',
                  'AGE':'edad',
                  'WITHDRAW_DATE':'fecha_baja_usuario',
                  'PREF_NAME':'pref_name_usuario',
                  'LATITUDE':'latitud_usuario',
                  'LONGITUDE':'longitud_usuario',
                  'CAPSULE_TEXT':'capsule_text_cupon',
                  'English_Translation_ct':'Cap_tex_traducido_cupon',
#                  'English_Translation_gn':'Genre_tex_traducido_cupon',
                  'GENRE_NAME':'genre_name_cupon',
                  'PRICE_RATE':'descuento_cupon',
                  'CATALOG_PRICE':'precio_full',
                  'DISCOUNT_PRICE':'precio_dto',
                  'DISPFROM':'fecha_inicio_disponible',
                  'DISPEND':'fecha_fin_disponible',
                  'DISPPERIOD':'dias_cupon_disponible',
                  'VALIDFROM':'fecha_inicio_valido',
                  'VALIDEND':'fecha_fin_valido',
                  'VALIDPERIOD':'dias_cupon_valido',
                  'USABLE_DATE_MON':'valido_lunes',
                  'USABLE_DATE_TUE':'valido_martes',
                  'USABLE_DATE_WED':'valido_miercoles',
                  'USABLE_DATE_THU':'valido_jueves',
                  'USABLE_DATE_FRI':'valido_viernes',
                  'USABLE_DATE_SAT':'valido_sabado',
                  'USABLE_DATE_SUN':'valido_domingo',
                  'USABLE_DATE_HOLIDAY':'valido_fiestas',
                  'USABLE_DATE_BEFORE_HOLIDAY':'valido_previo_fiestas',
                  'large_area_name':'large_area_name_cupon',
                  'ken_name':'ken_name_cupon',
                  'small_area_name':'small_area_name_cupon',
                  'COUPON_ID_hash':'ID_cupon_compra',
                  'ITEM_COUNT':'cantidad_cupon_compra',
                  'SALE_DATE':'fecha_compra',
                  'SMALL_AREA_NAME':'small_area_name_compra'},inplace=True)
```

```python
# Verificamos modificacion
ds.columns
```

```python
# Modificacion registros
ds['fecha_visita_dia'] = pd.to_datetime(ds['fecha_visita']).apply(lambda x: x.date())

ds['fecha_compra_dia'] = pd.to_datetime(ds['fecha_compra']).apply(lambda x: x.date())
```

```python
# Generamos archivo final en formato .CSV
nombre_archivo = 'ds_full'
ds.to_csv(ruta+nombre_archivo+extension,
         header=True,
         index=False)
```
